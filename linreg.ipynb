```python
import numpy as np
import random
from matplotlib import pyplot as plt
x=np.random.randn(1000)

y=(0.13*np.random.randn(1000))+x

w = np.random.randn(1)
b = np.random.randn(1)
learning_rate = 0.01
```

```python

for i in range(1000):
    
    
    y_pred = x * w + b
    cost = (1/(2*len(y))) * np.sum(np.square(y_pred - y)) 
    
    error = y - y_pred
    

    gradient_w = -2 * np.dot(error, x) / x.shape[0]
    gradient_b = -2 * np.sum(error) / x.shape[0]

    w -= learning_rate * gradient_w
    b -= learning_rate * gradient_b
    if i % 10 == 0:
        print("Cost at iteration", i, ":", cost)
    
```

```python
print("Weight:", w)
print("Bias:", b)
```

```python
from matplotlib import pyplot as plt

plt.scatter(x,y)
print(plt.plot(x,y_pred,c='red'))
```

```python
x=x.reshape(-1,1)
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
reg.fit(x,y)
print(reg.intercept_,"\n",reg.coef_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y, c='b')
plt.plot(x,y_pred, c='r')
```

As we can see both the methods output the same slope and intercept with very small differences at higher decimal places. This shows that gradient descent can find the global maxima of a convex function like linear regression. It reduces computational time and is suited for large datasets.
